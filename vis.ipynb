```python
import glob
import pandas as pd
import matplotlib.pyplot as plt

def plot_memory_usage():
    # Find all CSV files with the memory usage pattern
    csv_files = glob.glob('memory_usage_*.csv')

    plt.figure(figsize=(10, 6))  # Set the figure size for the plot

    # Process each file
    for file in csv_files:
        # Extract the label
        label = file.split('memory_usage_')[1].split('.csv')[0]

        # Read the data
        data = pd.read_csv(file)

        # Convert the 'timestamp' column to datetime format
        data['timestamp'] = pd.to_datetime(data['timestamp'])

        # Plot the data
        plt.plot(data['timestamp'], data['memory_usage_MB'], label=f'Process {label}')

    plt.xlabel('Time')
    plt.ylabel('Memory Usage (MB)')
    plt.title('Memory Usage Over Time')
    plt.legend()
    plt.tight_layout()  # Adjust layout to not overlap content
    plt.show()
```

```python
plot_memory_usage()
```
